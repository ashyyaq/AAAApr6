импортируем библиотеки

```python
import pandas as pd
import matplotlib.pyplot as plt
```

они импортировались

загружаем датасет и выводим его

```python
hsr = pd.read_csv(r"C:\Users\user\Downloads\hsr_character-data.csv")

hsr
```

собственно, датасет загрузился и вывелся

создаем линейные графики

```python
characters = hsr["character"].unique()[:3]

for character in characters:
    character_data = hsr[hsr["character"] == character]
    plt.plot(character_data["combat_type"], character_data["wb_ult"], label=character)

plt.xlabel("combat_type")
plt.ylabel("wb_ult")
plt.legend()
plt.show()
```

по идее должно было создаться 3 линейных графика с данными из столбца "character". они позволяют сравнить ульту разных персонажей в зависимости от их типа боя, ось x это тип боя, а ось y - значение ульты

создаем столбчатые диаграммы

```python
characters = hsr["character"].unique()[:3]

for character in characters:
    character_data = hsr[hsr["character"] == character]
    plt.bar(character_data["combat_type"], character_data["wb_b-atk"], label=character, alpha=0.5)

plt.xlabel("combat_type")
plt.ylabel("wb_b-atk")
plt.legend()
plt.show()
```

3 диаграммы создались, они позволяют сравнить базовую атаку персонажей с разными элементами

создаем круговые диаграммы

```python
combat_types = hsr["combat_type"].unique()
total_wb_b_atk = hsr.groupby("combat_type")["wb_b-atk"].sum()

plt.pie(total_wb_b_atk, labels=combat_types, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

здесь сначала группируются данные по типам боя и суммируем значения базовой атаки для каждого элемента. метод autopct='%1.1f%%' добавляет процентные значения на диаграмму

создаем  диаграммы рассеяния

```python
characters = hsr["character"].unique()[:3]

for character in characters:
    character_data = hsr[hsr["character"] == character]
    plt.scatter(character_data["wb_ult"], character_data["spd"], label=character)

plt.xlabel("wb_ult")
plt.ylabel("spd")
plt.legend()
plt.show()
```

создались 3 диаграммы рассеявания (дань хэн, март 7 и химеко), на них видна зависимость между ультой и скоростью персонажей

создаем гистограммы

```python
characters = hsr["character"].unique()[:5]

for character in characters:
    character_data = hsr[hsr["character"] == character]
    plt.hist(character_data["wb_b-atk"], label=character, alpha=0.5)

plt.xlabel("wb_b-atk")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

создались 5 гистограмм, на них представлены значения базовой атаки персонажей, позволяют сравнить распределение базовой атаки у персонажей
